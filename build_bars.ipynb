```python
import pandas as pd
df = pd.read_csv('../LOB Imbalance Alpha/data/processed.csv')
```

```python
df['timestamp'] = pd.to_datetime(df['event_time'], unit='ms')
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df = df.set_index('timestamp')
```

```python
bars = df['mid'].resample('1min').ohlc()
```

```python
bars.isna().sum()
```

```python
bars['close'].plot(title='BTCUSDT Mid Price (closing per minute)')
```
